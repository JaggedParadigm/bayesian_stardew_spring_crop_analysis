# Summary

This is the notebook accompanying an [article](https://www.cmshymansky.com/StardewSpringProfits/) I wrote about deciding the most profitable crop to plant in Spring of your 2nd year from a Bayesian perspective.

Here I demonstrate how to calculate the profit distribution for potatoes in Stardew Valley using these steps:

<ol>
  <li>
    Obtain items upon harvest distribution using corresponding counts and a Bayesian
    statistical technique
  </li>
  <li>
    Obtain quality distribution using a Bayesian statistical technique
  </li>
  <li>
    Obtain price distribution from the quality distribution and the corresponding quality prices
  </li>
  <li>
    Obtain the revenue per harvest distribution by combining the items per harvest and price
    distributions
  </li>
  <li>
    Calculate the number of harvests per year using using the days until first harvest
  </li>
  <li>
    Obtain the revenue per year distribution using the number of harvests and revenue per harvest
    distribution
  </li>
  <li>
    Calculate the expenses per year using the number of harvests per year and seed cost
  </li>
  <li>
    Obtain the distribution of profit per year using yearly expenses and the revenue per year
    distribution
  </li>
</ol>

# Obtain items upon harvest distribution

As mentioned in the main article, I'm assuming the developers determine the number of potatoes upon harvest by drawing a number between 0 and 1 repeatedly until it is above a certain hard-coded stopping value. The number of successes (i.e. where the number is below the threshold) is the number of extra potatoes drawn when pulling/harvesting a single plant.

To estimate the probability distribution of this extra-potato probability parameter I start with equal probability for 10,000 points between 0 and 1 (fractional version of probability) and update, according to Baye's Theorem, these probabilities with the probability of each item count upon harvest for that given parameter, assuming a geometric distribution modified to account for the guarunteed potato (i.e. the "success" probability exponent has a 1 subtracted from it).

Here's the corresponding code:

```python
import matplotlib.pylab as plt
import numpy as np
from functools import reduce

def pipe(data, *functions):
    return reduce(lambda a, x: x(a), functions, data)
def normalize_array(x):
    return x / x.sum()
def get_flat_prior(hypotheses):
    return {
        'hypothesis': hypotheses,
        'probability': (
            pipe(
                hypotheses,
                len,
                np.ones,
                normalize_array))}
def get_posterior_probility(prior_probability, likelihood_probability):
    return normalize_array(prior_probability * likelihood_probability)
def get_modified_geometric_probability(event_count, event_probability,
    guarunteed_count=0):
    """
    Probability event will occur event_count number of times, having already occurred 
    guarunteed_count number of times, until the event doesn't occur at the end given
    a fractional probability of occuring (event_probability)    
    """
    return (
        # Probability of failing
        event_probability**(event_count - guarunteed_count) 
        * (1 - event_probability))
def get_extra_item_draw_chance_distribution(item_counts, draw_chance_point_count=10000):
    return (
        pipe(
            {
                'point_count': draw_chance_point_count,
                'dataset': item_counts},
            lambda x: x | {
                'prior': (
                    get_flat_prior(
                        hypotheses=(
                            np.linspace(
                                0, 
                                1, 
                                x['point_count']))))},
            lambda x: x | {
                'likelihood_f_data': (
                    lambda data: (
                        get_modified_geometric_probability(
                            event_count=data, 
                            event_probability=x['prior']['hypothesis'],
                            guarunteed_count=1)))},
            lambda x: x | {
                'likelihood_probabilities': [
                    x['likelihood_f_data'](data)
                    for data in x['dataset']]},
            lambda x: x | {
                'posterior': {
                    'hypothesis': x['prior']['hypothesis'],
                    'probability': (
                        reduce(
                            get_posterior_probility, 
                            x['likelihood_probabilities'], 
                            x['prior']['probability']))}},
            lambda x: x['posterior']))
def line_plot_pmf(pmf, **plot_kwargs):
    return (
        pd.DataFrame(pmf)
        .set_index('hypothesis')
        ['probability']
        .plot(**plot_kwargs))
def get_cdf(pmf):
    return (
        pmf | {
            'probability': pmf['probability'].cumsum()})
def get_rough_percentile(cdf, percentile):
    """
    Rough because it uses the hypothesis at that index 
    directly instead of interpolating. Should be fine
    for the cdf of a pmf or for a finely-enough gridded 
    approximate pdf
    """
    return (
        pipe(
            np.where(cdf['probability'] > percentile)[0],
            lambda nearest_indices_2_percentile: (
                nearest_indices_2_percentile[0] if (
                    len(nearest_indices_2_percentile) > 0) else
                -1),
            lambda nearest_index_2_percentile: (
                cdf
                ['hypothesis']
                [nearest_index_2_percentile])))
def get_credible_interval(cdf, percentile_lower=0.03, percentile_upper=0.97):
    return {
        'lower_bound': get_rough_percentile(cdf, percentile_lower),
        'upper_bound': get_rough_percentile(cdf, percentile_upper)}
def get_maximum_a_posteriori(pmf):
    return (
        pd.DataFrame(pmf)
        .pipe(lambda x: x[x['probability'] == x['probability'].max()])
        ['hypothesis']
        .values
        [0])
def get_summary_statistics(pmf, percentile_lower=0.03, percentile_higher=0.97):
    return (
        pipe(
            pmf,
            get_cdf,
            get_credible_interval,
            lambda credible_interval: credible_interval | {
                'map': get_maximum_a_posteriori(pmf)}))
def get_summary_statistic_str(summary_statistics):
    return (
        f"{summary_statistics['map']:0.2f}"
        f" [{summary_statistics['lower_bound']:0.2f} "
        f"- {summary_statistics['upper_bound']:0.2f}]")
def get_pmf_summary_statistic_str(pmf):
    return (
        pipe(
            pmf,
            get_summary_statistics,
            get_summary_statistic_str))

potato_extra_item_draw_chance_distribution = (
    get_extra_item_draw_chance_distribution(
        item_counts=(
            potato_item_per_harvest_counts
            .pipe(lambda x: x[x['item'] == 'potato'])
            .pipe(
                lambda df: (
                    df
                    .loc[
                        df.index.repeat(df['count'])]
                    .reset_index(drop=True)))
            ['item_per_harvest']
            .values)))
(
    line_plot_pmf(
        potato_extra_item_draw_chance_distribution,
        title=(
            get_pmf_summary_statistic_str(
                pmf=potato_extra_item_draw_chance_distribution))))
plt.xlabel('extra potato draw chance')
_ = plt.ylabel('probability density')
```

The potato item harvest counts are consistent with an extra item draw chance between 11 and 32%, with the highest probability assigned to 20%.

This distribution is the probability of the extra potato draw chance parameter given the data. I now need the probability of future items upon harvest given the collected items upon harvest data. This is given by the posterior predictive distribution.

Here's the code I wrote to calculate it:

```python
def get_posterior_predictive(hypotheses, posterior, likelihood_f_parameters):
    return {
        'hypothesis': hypotheses,
        'probability': (
            np.sum(
                [
                    probability_of_probability * likelihood_f_parameters(probability)
                    for probability, probability_of_probability in (
                        zip(
                            posterior['hypothesis'],
                            posterior['probability']))],
                axis=0))}
def get_item_count_per_harvest_distribution(extra_item_draw_chance_distribution, counts):
    return (
        get_posterior_predictive(
            hypotheses=counts, 
            posterior=extra_item_draw_chance_distribution, 
            likelihood_f_parameters=(
                lambda parameters: (
                    get_modified_geometric_probability(
                        event_count=counts, 
                        event_probability=parameters,
                        guarunteed_count=1)))))
def barh_plot_pmf(pmf, **plot_kwargs):
    return (
        pd.DataFrame(pmf)
        .set_index('hypothesis')
        ['probability']
        .plot.barh(**plot_kwargs))


potato_item_count_per_harvest_distribution = (
    get_item_count_per_harvest_distribution(
        counts=np.arange(1, 11), 
        extra_item_draw_chance_distribution=potato_extra_item_draw_chance_distribution))
barh_plot_pmf(
    pmf=potato_item_count_per_harvest_distribution,
    title=(
        get_pmf_summary_statistic_str(
            pmf=potato_item_count_per_harvest_distribution)))
plt.xlabel('probability')
_ = plt.ylabel('potatoes upon harvest')
```

I used 10 potatoes upon harvest as the cut off since the probability for more is, essentially, zero.

The potato item harvest counts are consistent with drawing 1 to 3 potatoes per harvest.

# Obtain quality distribution

Here I derive the joint probability distribution for the chances of the 3 different qualities (i.e. regular, silver, gold) for potatoes. Note, that these quality probabilities themselves represent a probability distribution. Hence, the joint probability distribution is distribution of distributions!

I, again, start with a distribution with equal probabilities assigned to all possible quality distributions. I update these prior probabilities using a multinomial likelihood, which gives the probability of getting those quality counts given the counts and quality probabilities.

Here's the code where I obtain the resulting quality joint posterior:

```python
import itertools

def get_flat_3d_simplex_joint_prior(grid_point_count):
    return (
        pipe(
            {
                'hypothesis': (
                    np.array([
                        np.array(combination + (1.0 - sum(combination),))
                        for combination in (
                            list(
                                itertools.product(
                                    np.linspace(0, 1, grid_point_count),
                                    np.linspace(0, 1, grid_point_count))))
                        if sum(combination) <= 1]))},
            lambda x: x | {
                'probability': (
                    pipe(
                        x['hypothesis'].shape[0],
                        np.ones,
                        normalize_array))}))
def get_multinomial_probability(counts, probabilities):
    return np.prod(probabilities**counts, axis=1)
def get_item_per_harvest_distribution(quality_counts, parameter_point_count=100):
    return (
        pipe(
            {
                'point_count': parameter_point_count,
                'dataset': quality_counts},
            lambda x: x | {
                'prior': get_flat_3d_simplex_joint_prior(x['point_count'])},
            lambda x: x | {
                'likelihood': {
                    'hypothesis': x['prior']['hypothesis'],
                    'probability': (
                        get_multinomial_probability(
                            counts=x['dataset'], 
                            probabilities=x['prior']['hypothesis']))}},
            lambda x: x | {
                'posterior': {
                    'hypothesis': x['likelihood']['hypothesis'],
                    'probability': (
                        normalize_array(
                            x['prior']['probability'] 
                            * x['likelihood']['probability']))}},
            lambda x: x['posterior']))

potato_item_per_harvest_distribution = (
    get_item_per_harvest_distribution(
        quality_counts=potato_item_quality_counts['count'].values, 
        parameter_point_count=1000))
potato_item_per_harvest_distribution_df = (
    pd.DataFrame(
        np.column_stack([
            potato_item_per_harvest_distribution['hypothesis'],
            potato_item_per_harvest_distribution['probability']]),
        columns=['regular', 'silver', 'gold', 'probability']))
```

Here's a 3d plot of the quality distribution space:

```python
import plotly
import plotly.graph_objs as go

from plotly.graph_objs import (
    Scatter3d)

plot_figure = (
    go.Figure(
        data=[            
            go.Scatter3d(
                x=potato_item_per_harvest_distribution_df['regular'].values,
                y=potato_item_per_harvest_distribution_df['silver'].values,
                z=potato_item_per_harvest_distribution_df['gold'].values,
                mode='markers',
                marker={
                    'color': potato_item_per_harvest_distribution_df['probability'],
                    'size': 1,
                    'opacity': 0.8,
                    'colorscale':'Reds'})], 
        layout=(
            go.Layout(
                scene=go.Scene(
                    xaxis=go.XAxis(title='regular'),         
                    yaxis=go.YAxis(title='silver'),
                    zaxis=go.ZAxis(title='gold'))))))

# Render the plot.
plotly.offline.iplot(plot_figure)
```

Note that the possible quality combinations form a plane in 3 dimensions.

The darker the color the more probable the regular/silver/gold probability combinations.

Here are the probability distributions for the quality chances individually, from summing the probabilities where the quality draw chance is the same for each quality:

```python
marginal_distributions = (
    potato_item_per_harvest_distribution_df
    .set_index('probability')
    # Round to deal with numerical precision quirks
    .round(3)
    .reset_index()
    .pipe(
        lambda x: (
            pd.concat([
                x
                .groupby(quality)
                .sum()
                ['probability']
                .reset_index()
                .rename(columns={quality: 'hypothesis'})
                .assign(**{'quality': quality})
                for quality in ['regular', 'silver', 'gold']])
            .pivot(
                index='hypothesis',
                columns='quality',
                values='probability'))))
marginal_distributions.plot()

plt.xlabel('quality probability')
_ = plt.ylabel('probability density')
```

Regular looks more probable than silver which looks more probable than gold, which is what's expected.

Here I, again, use the posterior distribution to calculate the posterior predictive distribution representing the draw chance for each quality:

```python
potato_quality_distribution = (
    potato_item_per_harvest_distribution_df
    .assign(**{
        'regular': lambda x: x['probability'] * x['regular'],
        'silver': lambda x: x['probability'] * x['silver'],
        'gold': lambda x: x['probability'] * x['gold']})
    [['regular', 'silver', 'gold']]
    .sum()
    .reset_index()
    .pipe(lambda x: {
        'hypothesis': list(x['index'].values),
        'probability': x[0].values}))
barh_plot_pmf(potato_quality_distribution)
plt.xlabel('probability')
_ = plt.ylabel('quality')
```

This is the probability of different qualities given the already encountered quality counts.

# Obtain price distribution

Here I use the potato quality prices and the quality distribution to obtain the price distribution by simply replacing the qualities with their prices:

```python
potato_price_distribution = (
    potato_quality_distribution | {
        'hypothesis': (
            np.array([
                potato_item_properties_and_prices[quality].values[0]
                for quality in potato_quality_distribution['hypothesis']]))})
barh_plot_pmf(potato_price_distribution)
plt.xlabel('probability')
_ = plt.ylabel('potato price')
```

# Obtain revenue per harvest distribution

Now I need to combine the quality-based price distribution and the items upon harvest distribution to obtain that for the revenue. 

To do this, I need to build the revenue distribution for each potential item harvest count, then sum them but weighted by the probability of that count.

For example, the revenue distribution for 2 potatoes upon harvest can be built by summing all price combinations, multiplying the corresponding probabilities, then summing the probabilites for all duplicate sums.

For 3 items harvested, this procedure is performed for another potato price distribution and the distribution of revenue for 2 harvested potatoes. This can be performed as many times as the number of items harvested to obtain the full revenue distribution.

Here's the code I wrote to do it:

```python
def get_pmf_sum(pmf_1, pmf_2):
    return (
        pipe(
            [
                [
                    hypothesis_1 + hypothesis_2, 
                    probability_1 * probability_2]
                for hypothesis_1, probability_1 in (
                    zip(
                        pmf_1['hypothesis'], 
                        pmf_1['probability']))
                for hypothesis_2, probability_2 in (
                    zip(
                        pmf_2['hypothesis'], 
                        pmf_2['probability']))],
            lambda x: (
                reduce(
                    lambda a, y: a | {y[0]: a.get(y[0], 0.0) + y[1]},
                    x,
                    {})),
            lambda x: {
                'hypothesis': np.array(list(x.keys())), 
                'probability': np.array(list(x.values()))}))
def get_multiple_pmf_sums(pmf, sum_count):
    return (
        reduce(
            get_pmf_sum,
            sum_count * [pmf]))
def get_revenue_pmf(count_pmf, price_pmf):
    return (
        pipe(
            [
                pipe(
                    get_multiple_pmf_sums(
                        price_pmf, 
                        count),
                    lambda x: x | {'probability': probability * x['probability']},
                    lambda x: dict(zip(x['hypothesis'], x['probability'])))
                for count, probability in zip(
                    count_pmf['hypothesis'],
                    count_pmf['probability'])],
            lambda x: (
                reduce(
                    lambda a, y: {
                        key: a.get(key, 0.0) + y.get(key, 0.0)
                        for key in set(itertools.chain(a, y))},
                    x,
                    {})),
            lambda x: (
                (
                    pd.DataFrame(
                        x.items(), 
                        columns=['hypothesis', 'probability'])
                    .sort_values('hypothesis')
                    .pipe(lambda x: {
                        'hypothesis': x['hypothesis'].values,
                        'probability': x['probability'].values})))))

potato_revenue_per_harvest_distribution = (
    get_revenue_pmf(
        count_pmf=potato_item_count_per_harvest_distribution, 
        price_pmf=potato_price_distribution))
(
    pd.DataFrame(potato_revenue_per_harvest_distribution)
    .pipe(lambda x: x[x['hypothesis'] <= 520])
    .set_index('hypothesis')
    ['probability']
    .plot.barh()
)
plt.xlabel('probability')
_ = plt.ylabel('potato revenue per harvest')
```

The maximum of this distribution is actually 3600 gold for 10 potatoes. I chose that y-axis cut off because the probabilities after were so small.

# Obtain number of harvests per year

Here I use the time until first harvest to estimate the number of harvests per year for potatoes:

```python
def get_harvest_count(days_2_first_harvest, 
    days_in_period=28, days_2_next_harvest=None, 
    not_planted_already=True):
    return (
        pipe(
            (
                days_2_first_harvest + 1 if not_planted_already else
                0),
            lambda initial_plant_and_mature_window: (
                days_in_period 
                - initial_plant_and_mature_window),            
            lambda days_left: (
                pipe(
                    1 
                    + (
                        days_2_first_harvest if days_2_next_harvest is None else
                        days_2_next_harvest),
                    lambda harvest_plant_and_mature_window: (
                        # Number of harvests from full harvest/plant/mature windows
                        days_left // harvest_plant_and_mature_window 
                        # Extra harvest from first day of final partial window
                        + (
                            0 if days_in_period % harvest_plant_and_mature_window == 0 else 
                            1))))))
potato_yearly_harvest_count = (
    get_harvest_count(
        days_2_first_harvest=(
            potato_item_properties_and_prices
            ['days_til_first_harvest']
            .values
            [0])))
potato_yearly_harvest_count
```

Turns out potatoes can only be planted 3 times a year in the same square given the time until first harvest and the availble growing window.

# Obtain revenue per year distribution

Now I use the number of yearly harvests and the revenue per harvest distribution to calculate the yearly revenue distribution for potatoes:

```python
potato_yearly_revenue_distribution = (
    get_multiple_pmf_sums(
        pmf=potato_revenue_per_harvest_distribution, 
        sum_count=potato_yearly_harvest_count))
(
    pd.DataFrame(potato_yearly_revenue_distribution)
    .pipe(lambda x: x[x['hypothesis'] <= 1400])
    .set_index('hypothesis')
    .sort_index()
    ['probability']
    .plot.barh(figsize=(10, 10))
)
plt.xlabel('probability')
_ = plt.ylabel('potato yearly revenue')
```

I cut off the figure at 1400 gold because it goes all the way to 10800 (10 gold potatoes for all 3 harvests)!

# Obtain expenses per year

The expenses are just the number of harvests multiplied by the seed cost:

```python
potato_yearly_expenses = (
    potato_item_properties_and_prices
    .assign(**{
        'yearly_expenses': lambda x: (
            potato_yearly_harvest_count
            * x['initial_investment'])})
    ['yearly_expenses']
    .values
    [0])
potato_yearly_expenses
```

# Obtain profits per year distribution

We just need to subtract the expenses from the yearly revenues of the yearly revenue distribution to obtain the profit distribution:

```python
potato_yearly_profit_distribution = (
    potato_yearly_revenue_distribution | {
        'hypothesis': (
            potato_yearly_revenue_distribution['hypothesis'] 
            - potato_yearly_expenses)})

(
    pd.DataFrame(potato_yearly_profit_distribution)
    .pipe(lambda x: x[x['hypothesis'] <= 1400])
    .set_index('hypothesis')
    .sort_index()
    ['probability']
    .plot.barh(figsize=(10, 10))
)
plt.xlabel('probability')
_ = plt.ylabel('potato yearly profit')
```

There we have it!
